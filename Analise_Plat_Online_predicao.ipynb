```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
base_mkt = pd.read_csv("/content/MKT.csv")
```

```python
sns.pairplot(base_mkt);
```

```python
sns.pairplot(base_mkt, x_vars=["youtube",
                               "facebook", "newspaper"], y_vars = "sales")
```

```python
x = base_mkt[["youtube", "facebook", "newspaper"]]
y = base_mkt["sales"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Erro Quadrático Médio (MSE): {mse}')
print(f'Coeficiente de Determinação (R²): {r2}')
```

```python
plt.scatter(x_test['youtube'], y_test, color='blue', label='Atual')
plt.scatter(x_test['youtube'], y_pred, color='red', label='Previsão')
plt.xlabel('youtube')
plt.ylabel('sales')
plt.title('youtube VS sales')
plt.legend()
plt.show()
```

```python
plt.scatter(x_test['facebook'], y_test, color='blue', label='Atual')
plt.scatter(x_test['facebook'], y_pred, color='red', label='Previsão')
plt.xlabel('facebook')
plt.ylabel('sales')
plt.title('facebook VS sales')
plt.legend()
plt.show()
```

```python
plt.scatter(x_test['newspaper'], y_test, color='blue', label='Atual')
plt.scatter(x_test['newspaper'], y_pred, color='red', label='Previsão')
plt.xlabel('newspaper')
plt.ylabel('sales')
plt.title('newspaper VS sales')
plt.legend()
plt.show()
```
